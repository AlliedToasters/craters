```python
from proposal_code import tiles
from test_human import invert_image
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
import time
import cv2
```

```python
img = np.array(tiles['3_24'])
inv_img = invert_image(img)
```

```python
plt.imshow(img, cmap='Greys');
```

```python
plt.imshow(inv_img, cmap='Greys');
```

```python
def med_filt(input_array):
    """Completes the median filter-subtract from image step."""
    return input_array-ndimage.median_filter(input_array, 4)

t1 = time.time()
filt1 = med_filt(img)
plt.imshow(filt1, cmap='Greys');
print('filter applied in {} seconds.'.format(time.time() - t1))
```

```python
t1 = time.time()
inv_filt1 = med_filt(inv_img)
plt.imshow(inv_filt1, cmap='Greys');
print('filter applied in {} seconds.'.format(time.time() - t1))
```
