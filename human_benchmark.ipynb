```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
import test_human
%matplotlib inline
```

```python
df = pd.read_csv('first_attempt.csv')
```

```python
df.head()
```

```python
Y_ = df.prediction
Y = df.crater
metrics.classification_report(Y, Y_)
```

```python
fp = np.where((df.crater==0) & (df.prediction==1), True, False)
fn = np.where((df.crater==1) & (df.prediction==0), True, False)
```

```python
print('{} false positives.'.format(len(df[fp])))
print('{} false negatives.'.format(len(df[fn])))
```

```python
for i, row in df[fp].iterrows():
    ID = row['id']
    print('false positive example: ')
    test_human.show_example(ID)
```

```python
for i, row in df[fn].iterrows():
    ID = row['id']
    print('false negative example: ')
    test_human.show_example(ID)
```
